# Set Operations

# Union-> Combines elemnets from teo sets excluding duplictes.
# intersection-> only common element between sets.
# Difference-> return the element that is present in first  set and not in second.
# Symmetric Difference-> returns element thta are present in either of the sets 

```python
s1={1,2,3,4}
s2={5,6,7,8,2,3,4}
```

```python
s1|s2 #union
```

```python
s1 &s2 #intersection
```

```python
s1-s2 #Difference
```

```python
# symmetric diff
s1^s2
```
