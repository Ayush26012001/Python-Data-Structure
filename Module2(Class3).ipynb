# List Comprehension
- Writing consize code

```python
p=[10,20,30,40,50]
d=[]
for i in p:
    d.append(i*2)
print(d)
```

```python
[i*2 for i in p]
```

```python
names =['ayush',"rahul",'vijay']
[i.capitalize() for i in names]
```

```python
a='ayush'
a.capitalize()
```

```python
num=[1,2,3,4,5]
[i**2 for i in num]
```

```python
string1='doc.ppt'
string1.split(".")[-1]
```

```python
file_name=['doc.ppt','doc1.pdf','doc2.jpg','doc.py']
[file.split(".")[-1] for file in file_name]
```

```python
#conditional list comprehensio
email_add=["ah@gamil.com",'sj@yahoo.com','mj@yahoo.com','abc@gmail.com']
[i for i in email_add if i.endswith('@yahoo.com')]
```

```python
# Nested lis comprehension
pairs=[]
for x in [1,2,3,4]:
    for y in [5,6,7,9]:
        pairs.append([x,y])
print(pairs)
```

```python
[[x,y] for x in [1,2,3] for y in [4,5,6]]
```
