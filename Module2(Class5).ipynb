# Dictionary Comprehension


```python
# Zip function
stu=['a','b','c']
marks=[80,95,100]
for i in zip(stu,marks):
    print(i)
```

```python
# storing aboabe info in dictionary
stu=['a','b','c']
marks=[80,95,100]
stu_dict={}
for stu,marks in zip(stu,marks):
    stu_dict[stu]=marks
stu_dict
```

```python
# Dictionary Comprehension
stu=['a','b','c']
marks=[80,95,100]
{student:marks for student, marks in zip(stu,marks)}
```

```python
id=[1,2,3]
name=['a1','s2','s3']
{i:n for i,n in zip(id,name)}
```
